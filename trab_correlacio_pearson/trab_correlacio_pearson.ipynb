```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
personas = pd.read_csv("../datos/personas.csv")
personas
```

```python
# Graficos de dispersion
plt.scatter(personas["altura"], personas["peso"])
plt.xlabel("Altura (cms)")
plt.ylabel("Peso (Kgs)")
plt.show()
```

Cada punto asumimos que es una perosna, donde relacionamos con su altura y su peso

```python
# Relacion: Ingreso y horas trabajadas
plt.scatter(personas["ingreso"], personas["horas_trabajadas"])
plt.xlabel("Ingreso ($)")
plt.ylabel("Horas trabajadas)")
plt.show()
```

```python
# Relacion: Ingreso y Ausencias
plt.scatter(personas["ingreso"], personas["ausencias"])
plt.xlabel("Ingreso ($)")
plt.ylabel("Ausencias)")
plt.show()
```

```python
# Relacion: Ingreso y Peso podrian no ajustarse a una linea recta
plt.scatter(personas["ingreso"], personas["peso"])
plt.xlabel("Ingreso ($)")
plt.ylabel("Peso (kgs))")
plt.show()
```

```python
# Matriz correlaciones
# -1 (Existe una relacion fuerte entre las 2 variables)
# +1 (Existe una relacion fuerte entre las 2 variables)
# el signo nos indica la direccion
personas.corr()
```

```python
matriz = personas.corr()
plt.matshow(matriz, cmap="bwr")
plt.xticks(range(5), personas.columns, rotation = 90)
plt.yticks(range(5), personas.columns)
for i in range(len(personas.columns)):
    for j in range(len(personas.columns)):
        plt.text(i,j, round(matriz.iloc[i,j],2), ha="center", va="center")
plt.colorbar()
plt.show()
```
