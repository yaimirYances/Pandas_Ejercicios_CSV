```python
import pandas as pd
import numpy as np
```

```python
peliculas = pd.read_csv("datosnum.csv")
df = pd.DataFrame(peliculas)
df
```

```python
import datetime
hora = df['A'].str.extract('((?:[01]\d|2[0-3]):[0-5]\d:[0-5]\d)')
hora
```

```python
#USANDO EPRESIONES REGULARES PARA EXTRAER SOLO LOS NUMEROS
df2 = pd.DataFrame()
df2["A"]=df['A'].str.extract('((?:[01]\d|2[0-3]):[0-5]\d:[0-5]\d)')
df2["C"]=df['C'].str.extract('((?:[01]\d|2[0-3]):[0-5]\d:[0-5]\d)')
df2["F"]=df['F'].str.extract('(\d+(?:\.\d+)?)')
df2["G"]=df['G'].str.extract('(\d+(?:\.\d+)?)')
df2
```

```python
#AGRUPAR POR MINUTOS
df2["A"] = pd.DatetimeIndex(df2["A"])
df2.set_index(keys="A", inplace=True)
inicio = datetime.time(11,18,0)
fin = datetime.time(11,19,0)
```

```python
#MOSTRAR LOS DATOS ENTRE LAS 2 FECHAS
df2[["F"]].between_time(inicio,fin)
```
