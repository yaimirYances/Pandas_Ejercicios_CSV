```python
#pip install fuzzywuzzy
import pandas as pd
from fuzzywuzzy import process
```

```python
carga = pd.read_csv("wuzy.csv")
df = pd.DataFrame(carga)
df
```

```python
ok = {"Mexico","Argentina","Chile","Ecuador","Colombia"}
usuario = "america mexico"
aprox = process.extractOne(usuario,ok)
aprox[1]
```

```python
nombres = []
proporciones = []
def cambiosCorrectos(mal,bien):
    for filas in mal:
        x = process.extractOne(filas,bien)
        nombres.append(x[0])
        proporciones.append(x[1])
    return nombres, proporciones
```

```python
mal = df["mal"].dropna().values
bien = df["bien"].dropna().values
```

```python
corregido, rango = cambiosCorrectos(mal, bien)
df["correccion"]=pd.Series(corregido)
df["ration"]=pd.Series(rango)
df
```
