```python
import pandas as pd
import numpy as np
```

```python
minutos = pd.read_csv("minutos2.csv")
df = pd.DataFrame(minutos)
df
```

```python
alto = df.loc[0,"B"]
for i in range(1,len(df)):
    if df.loc[i,"B"]>df.loc[i-1,"B"] and df.loc[i,"B"]>alto:
        df.loc[i,"DB"]=df.loc[i,"B"]-alto
        alto = df.loc[i,"B"]
    else:
        df.loc[i,"DB"]=0
```
