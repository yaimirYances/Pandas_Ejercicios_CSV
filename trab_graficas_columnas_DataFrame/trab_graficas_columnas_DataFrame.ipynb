```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
carga = pd.read_csv("VGMod.csv")
df = pd.DataFrame(carga)
df
```

```python
colors  = ("dodgerblue","salmon", "palevioletred", 
           "steelblue", "seagreen", "plum", 
           "blue", "indigo", "beige", "yellow")

i=0
for col in df:
    sizes = df[col].value_counts()
    # pie = df[col].value_counts().plot(kind="pie", colors=colors, shadow=True, autopct="%1.1f%%", 
    #                                   labels=None,startangle=30, radius=1.5, center=(0.5,0.5), textprops={"fontsize":12}, frame=False, pctdistance=1.3)
    
    pie =df[col].value_counts().plot(kind="bar")
    labels = sizes.index.unique()
    #plt.gca().axis("equal")
    plt.title(df.columns[i], weight="bold", size=14)
    plt.subplots_adjust(left=0.0, bottom=0.1, right=0.85)
    plt.savefig(str(df.columns[i])+".png", dpi=100, bbox_inches="tight")
    pie.set_ylabel("")
    #plt.legend(labels, bbox_to_anchor=(0.5,-.2))
    i=i+1
    plt.show()
    
```
