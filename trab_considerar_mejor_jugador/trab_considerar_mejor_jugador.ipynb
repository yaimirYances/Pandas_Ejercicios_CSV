```python
import pandas as pd
```

```python
carga = pd.read_csv("archive/FIFA21_official_data.csv")
df = pd.DataFrame(carga)
df
```

```python
# GRAFICA DE LOS POLIGONOS
df1 = df[df['Best Overall Rating']>=90]
len(df1)
```

```python
df2 = df1.head(5)
df2
```

```python
columnas = ["Name","BallControl","Stamina","Strength","Jumping"]
df2 = df2[[*columnas]]
df2.head(5)
```

```python
import numpy as np
```

```python
angulos = [i / float(len(df2)) * 2 * np.pi for i in range(len(df2))]
angulos += angulos[:1]
np.degrees(angulos)
```

```python
df2 = df2._append(df2.iloc[0])
df2
```

```python
df3 = df2.drop(["Name"], axis=1)
df3
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.subplot(polar=True)
plt.plot(angulos, df2["BallControl"])
plt.fill(angulos, df2["BallControl"], "b", alpha=0.1)

plt.plot(angulos, df2["Stamina"])
plt.fill(angulos, df2["Stamina"], "g", alpha=0.1)

plt.plot(angulos, df2["Strength"])
plt.fill(angulos, df2["Strength"], "r", alpha=0.1)

plt.plot(angulos, df2["Jumping"])
plt.fill(angulos, df2["Jumping"], "o", alpha=0.1)

#plt.xticks(angulos[:-1], df2["Name"])

plt.legend(labels=(df3.columns), loc=1)
plt.title("FIFA 21")
```
