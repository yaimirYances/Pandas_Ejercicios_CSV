```python
import pandas as pd
import numpy as np
```

```python
minutos = pd.read_csv("FIFA21_official_data.csv")
df = pd.DataFrame(minutos)
df
```

```python
df.dtypes
```

```python
df.info()
```

```python
# BUSQUEDAS SIMILARES
# case=True no distinguir mayusculas y minisculas
df[df["Club"].str.contains("madrid|barcelona|city", case=False, na=False, regex=True)]
```

```python
jugadores = list(df["Name"].head(10))

listar = "|".join(jugadores)
df[df["Name"].str.contains(listar, case=False, na=False, regex=True)]
```

```python
# BUSQUEDAS EXACTAS
df["Name"]=df["Name"].str.strip() # quitar espacios en blancos
df[df["Name"].str.match("M. Neuer", case=False, na=False)]
```

```python
jugadores = list(df["Name"].head(10))
listar = "|".join(jugadores)
df[df["Name"].str.match(listar, case=False, na=False)]
```

```python
# BUSQUEDA COMIENZA CON
df[df["Nationality"].str.startswith("Col")]
```

```python
df["Nationality"]=df["Nationality"].str.strip() # quitar espacios en blancos
```

```python
inicia_con = ["C","M"]
nacionalidades = "|".join(inicia_con)
df[df["Nationality"].str.startswith(tuple(nacionalidades))]
```

```python
# BUSQUEDA FINALIZA CON
df[df["Name"].str.endswith("z")]
```
